```python

urllink = "https://pureportal.coventry.ac.uk/en/organisations/research-centre-for-computational-science-and-mathematical-modell/publications/"

publication_details = []
authors_names = []

#Count of url containing publications (1 from url1)
total_urls = 1

#Count of total publications
total_pubs = 0

# Count of publications with at least one CSM member
pub_csm = 0


while urllink:
    response = requests.get(urllink)
    soup = BeautifulSoup(response.text, "html.parser")
    articles = soup.find_all("div", class_="result-container")

    for article in articles:
        total_pubs = total_pubs + 1
        # Extract details for each publication and add to publication_details list
        title_elem = article.find("h3", class_="title")
        title = title_elem.text.strip()
        link = title_elem.find("a").get("href")
        date_elem = article.find("span", class_="date")
        date = date_elem.text.strip()
        author_elems = article.find_all("a", class_="link person")
        authors = []
        for author_elem in author_elems:
            author = author_elem.text.strip()
            authors_names.append(author)
            author_data = {"name": author, "href": author_elem.get("href")}
            authors.append(author_data)
        if authors:
            publication_details.append({"title": title, "link": link, "date": date, "authors": authors})
            pub_csm = pub_csm + 1

    # Get the link to the next page of results, if it exists
    next_page = soup.find("li", class_="next")
    if next_page:
        total_urls = total_urls + 1
        next_page_link = urljoin(urllink, next_page.a["href"])
        urllink = next_page_link
    else:
        urllink = None

    # Wait for 5 seconds before making the next request
    time.sleep(5)

#print(publication_details)
print("Total url for publication:", total_urls)
print("Total publications on the CSM website is:", total_pubs)
print("Total publications with at one lest CSM member as author:", pub_csm)


with open("data.json", "w") as f:
        json.dump(publication_details, f, indent=4)


```

```python
import json
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')

# Define the Porter stemmer and the stop words
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

all_tokens = []

# Load the publication details from the JSON file
with open("data.json", "r") as f:
    publication_details = json.load(f)

# Preprocess each publication
for i,pub in enumerate(publication_details):
    # Tokenize the title
    title_tokens = word_tokenize(pub['title'])

    # Tokenize the authors
    author_tokens = []
    for author in pub['authors']:
        author_tokens.extend(word_tokenize(author['name']))

    # Combine the tokens
    doc_tokens = title_tokens + author_tokens

    # Remove stop words and perform stemming
    filtered_tokens = []
    for token in doc_tokens:
        if token.lower() not in stop_words:
            stemmed_token = stemmer.stem(token.lower())
            filtered_tokens.append(stemmed_token)

    # Add the filtered tokens to the list of all tokens
    all_tokens.extend(filtered_tokens)

    # Update the publication details with the filtered tokens
    publication_details[i]['tokens'] = filtered_tokens



# Print the list of all tokens
print(all_tokens)
print("*" *50)
print(publication_details)

# Save the preprocessed publication details to a JSON file
with open("preprocessed_data.json", "w") as f:
    json.dump(publication_details, f, indent=4)

print("Preprocessing complete.")
```

```python
import json
from collections import defaultdict
from math import log

# Load the preprocessed publication details from the JSON file
with open("preprocessed_data.json", "r") as f:
    publication_details = json.load(f)

#print(publication_details)
# Define a function to compute the inverse document frequency (IDF)
def compute_idf(token, all_tokens):
    N = len(publication_details)
    df = sum(1 for tokens in all_tokens.values() if token in tokens)
    idf = log(N / df)
    #print(token)
    #print(idf)
    return idf

# Create a dictionary to store the token frequency (TF) for each document
tf = defaultdict(dict)
for i, pub in enumerate(publication_details):
    for token in pub['tokens']:
        if token not in stop_words:
            tf[i][token] = tf[i].get(token, 0) + 1

# Create a dictionary to store the inverted index
inverted_index = defaultdict(list)
for i, pub in enumerate(publication_details):
    for token in set(pub['tokens']):
        if token not in stop_words:
            tfidf = tf[i][token] * compute_idf(token, tf)
            inverted_index[token].append((i, tfidf))

# Compute the TF-IDF weights for each token in each document
tfidf = []
for i, pub in enumerate(publication_details):
    for token in set(pub['tokens']):
        if token not in stop_words:
            tfidf_weight = tf[i][token] * compute_idf(token, tf)
            tfidf.append({'doc_id': i, 'token': token, 'tfidf': tfidf_weight})

# Print the top 10 TF-IDF weights for each document
for i, pub in enumerate(publication_details):
    print(f'Document {i} - {pub["title"]}')
    doc_tfidf = [t for t in tfidf if t['doc_id'] == i]
    doc_tfidf.sort(key=lambda x: x['tfidf'], reverse=True)
    for j in range(min(10, len(doc_tfidf))):
        print(f'   {doc_tfidf[j]["token"]}: {doc_tfidf[j]["tfidf"]}')

# Save the inverted index and TF-IDF weights to JSON files
with open("inverted_index.json", "w") as f:
    json.dump(inverted_index, f, indent=4)

with open("tfidf.json", "w") as f:
    json.dump(tfidf, f, indent=4)
```
