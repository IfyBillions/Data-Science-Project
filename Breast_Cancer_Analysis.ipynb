```python
##creating a balance in the target variable
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
## separate the dataset into X and y
X = breast_cancer.drop("Status", axis =1)
y = breast_cancer["Status"].values
```

```python
X
```

```python
y
```

```python
## applying SMOTE balancing technique
sm = SMOTE(sampling_strategy="auto",random_state=42)
X_res, y_res = sm.fit_resample(X,y)

## converting the resample arrays into a dataframe
# Convert the resampled arrays back to a DataFrame
breast_cancer_resample = pd.DataFrame(X_res, columns=X.columns)
breast_cancer_resample['Status'] = y_res

# Check the distribution of the target variable after resampling
print("Original target variable distribution:", Counter(y))
print("Resampled target variable distribution:", Counter(y_res))

## visualizing the resampled data
sns.countplot(x='Status',data=breast_cancer_resample, palette='gray')
plt.title('Balanced Target Variable Distribution')
plt.show()
```

```python
# Feature scaling of the data using Standard Scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_res)

# convert scaled features back to a dataFrame
breast_cancer_scaled = pd.DataFrame(X_scaled, columns=X.columns)
breast_cancer_scaled['Status'] = y_res
```

```python
print(breast_cancer_scaled.head())
print(X_res, len(X_res))
print(y_res, len(y_res))
```

```python
## Splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_res, test_size=0.2)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
## predicting on the test
y_pred = rf.predict(X_test)

## Evaluate with different metrics
print("Random Forest:")
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted Alive', 'Predicated Dead'],yticklabels=['Actual Alive','Actual Dead'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
